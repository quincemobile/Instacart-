```python
#this is a mess so i'm going to aggregate the data into age groups
```

```python
# Define the bins for age groups
bins = [0, 19, 29, 39, 49, 59, 69, 79, 999]  # 999 is an arbitrary large number for 80 and up
labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']
```

```python
# Create a new column for age groups
df_data['age_group'] = pd.cut(df_data['Age'], bins=bins, labels=labels, right=True)
```

```python
# Calculate median income per age group
age_group_income = df_data.groupby('age_group')['income'].median().reset_index()
```

```python

# Explicitly create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

```

```python
# Use the axes 'ax' for the scatter plot
sns.scatterplot(x='age_group', y='income', data=age_group_income, ax=ax, s=100)

```

```python
# Save the figure that contains this specific scatter plot
file_path = os.path.join(main_dir_path, '04 Analysis', 'Visualizations', 'scatterplot_income_age_group.png')
fig.savefig(file_path)
```

```python
# Plotting the aggregated data
plt.figure(figsize=(10, 6))  # You can adjust the figure size as needed
scatter_plot = sns.scatterplot(x='age_group', y='income', data=age_group_income, s=100)  # s is the marker size
```

```python
#given the scatterplot above, we can see that income rises sharply in the 40 and up age groups from 70k or less under 40 to over 110k after 40.
```
